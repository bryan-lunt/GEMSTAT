```python
x = np.linspace(0.0,1.0,100)

gradient_1 = np.linspace(1.0,0.0,100)
gradient_2 = np.linspace(0.0,1.0,100)

inputs = GSMAT.GEMSTAT_Matrix()
inputs.add_row("first",gradient_1)
inputs.add_row("second",gradient_2)

inputs.write("./factor_expr.tab")
```

```python
def hill_function(concentration, half_at=0.5,coop=1.0):
    return S.power(S.power(half_at/concentration,coop)+1.0,-1.0)
```

```python
import pylab
pylab.plot(hill_function(gradient_1,0.5,1.0) )
pylab.ylim(0.0,1.0)
```

```python
a = hill_function(gradient_1,0.5,1.0)
b = 2.0*hill_function(gradient_1,0.5,1.0)
c = b
d = b

expr_1 = GSMAT.GEMSTAT_Matrix()
expr_1.add_row("a",hill_function(gradient_1,0.5,1.0))
expr_1.add_row("b",2.0*hill_function(gradient_1,0.5,1.0))
expr_1.add_row("c",2.0*hill_function(gradient_1,0.5,1.0))
expr_1.add_row("d",2.0*hill_function(gradient_1,0.5,1.0))
expr_1.write("./expr.1")
```
